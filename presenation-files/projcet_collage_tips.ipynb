# reviewing google reviews 

```python
# google reviews
df_google_review.head(10)
```

```python
df_google_review.head()
```

```python
# selecting reviews and name giveing a postive or negitve resopnce using pretrained supervisered ml models
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline

# load per-trained model

model_name = "nlptown/bert-base-multilingual-uncased-sentiment"
model = AutoModelForSequenceClassification.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
print('models imported')
#creating inference pipeline
classifier = pipeline('sentiment-analysis',model = model , tokenizer = tokenizer)
print('classifier imported')

# predict scores
reviews = df_google_review['text'].tolist()
result = classifier(reviews)
print('stated_reviews')

# scoreing 
scores = [int(result['label'].split()[0]) for result in result]
print('impementing scores to df ')

#adding predicted score
df_google_review['predicted_socre'] = scores
```

```python
df_google_review.head(10)
```

```python
plt.plot(df_google_review['predicted_socre'])
plt.xlabel("totalScore")
plt.ylabel("predicted_socre")
plt.show()
```

```python
#adding reating to the comments 
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline

# load per-trained model

model_name = "nlptown/bert-base-multilingual-uncased-sentiment"
model = AutoModelForSequenceClassification.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
print('models imported')
#creating inference pipeline
classifier = pipeline('sentiment-analysis',model = model , tokenizer = tokenizer)
print('classifier imported')

# predict scores
df_youtube = df_youtube[df_youtube['comment'].notna()]
reviews = df_youtube['comment'].astype(str).tolist()     # Ensure strings
result = classifier(reviews)
print('stated_reviews')

# scoreing 
scores = [int(result['label'].split()[0]) for result in result]
print('impementing scores to df ')

#adding predicted score
df_youtube['predicted_socre'] = scores
print("added to data_frame")
```
